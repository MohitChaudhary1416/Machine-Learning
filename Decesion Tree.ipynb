```python
# Decesion tree
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
x = data.drop('species', axis = 1)
y = data.species
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size = 0.2, random_state = 12)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dmodel = DecisionTreeClassifier()
dmodel.fit(xtrain,ytrain)
```

```python
# Prediction
# Evaluation
# View Tree
```

```python
dmodel.predict([[5.2,3,2.5,1]])
```

```python
# Evaluation
yprediction = dmodel.predict(xtest)
yprediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(ytest,yprediction)
print(f'Accuracy = {accuracy:.2%}')
```

```python
from sklearn.tree import plot_tree
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))  # Set figure size first
plot_tree(dmodel, filled=True, feature_names=x.columns, class_names=y.unique().astype(str))
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Diabetes.csv')
df.head()
```

```python
x = df[['Glucose', 'Age']]
y = df.Diabetes
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y , test_size = 0.2, random_state = 5)
```

```python
model = DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
yprediction1 = model.predict(xtest)
yprediction1
```

```python
accuracy1 = accuracy_score(ytest,yprediction1)
print(f'Accuracy = {accuracy1:.2%}')
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))  # Set figure size first
plot_tree(model, filled=True, feature_names=x.columns, class_names=y.unique().astype(str))
plt.show()
```

```python
# Testing Accuracy using Training Data
```

```python
ypred_train = model.predict(xtrain)
```

```python
acc_train = accuracy_score(ytrain, ypred_train)
print(f'Accuracy = {acc_train:.2%}')
```

```python
# Low accuracy in testing
# High accuracy in training:=> Overfitting
```

```python
# To see Parameters
model.get_params()
```

```python
model2 = DecisionTreeClassifier(max_depth = 4,min_samples_leaf = 2)
model2.fit(xtrain, ytrain)
```

```python
ypredi = model2.predict(xtest)
```

```python
accuracy2 = accuracy_score(ytest,ypredi)
print(f'Accuracy = {accuracy2:.2%}')
```

```python
from sklearn.model_selection import GridSearchCV
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data1 = pd.read_csv('Diabetes.csv')
data1.head()
```

```python
x = data1[['Glucose', 'Age']]
y = data1.Diabetes
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2, random_state = 5)
```

```python
parameter = {
    'max_depth':[4,5,6,7],
    'max_leaf_nodes':[4,5,6]
}
```

```python
grid_search = GridSearchCV(
    estimator = DecisionTreeClassifier(),
    param_grid = parameter,
    scoring = 'accuracy'
)
grid_search.fit(xtrain, ytrain)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
grid_search.predict([[85,31]])
```

```python
# Testing data
yprediction = grid_search.predict(xtest)
```

```python
accuracy = accuracy_score(ytest,yprediction)
print(f'ACCURACY = {accuracy:.2%}')
```
