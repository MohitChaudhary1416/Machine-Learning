## Support Vector Machines 

```python
import pandas as pd
from sklearn.svm import SVC 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples = 200, n_features = 2, centers = 2, cluster_std = 0.15, random_state = 5 )
```

```python
x_df = pd.DataFrame(X, columns = ['A', 'B']) # Changing array into dataframe
x_df
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data = x_df, x = 'A', y = 'B' ,hue = y)
plt.show()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x_df,y,test_size = 0.2, random_state = 7)
```

```python
svm = SVC()
svm.fit(xtrain,ytrain)
```

```python
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(200,noise=0.05,random_state = 5)
```

```python
x_df = pd.DataFrame(X, columns = ['A','B'])
x_df
```

```python
sns.scatterplot(data = x_df, x = 'A', y = 'B', hue = y)
plt.show()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x_df, y ,test_size = 0.2, random_state = 7)
```

```python
svm = SVC(kernel ='rbf')
svm.fit(xtrain,ytrain)
```

```python
plot_decision_regions(X = X, y =y , clf = svm)
```

```python
ypred = svm.predict(xtest)
```

```python
accuracy = accuracy_score(ytest,ypred)
print(f'ACCURACY = {accuracy:.2%}')
```
