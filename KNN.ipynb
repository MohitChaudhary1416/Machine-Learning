```python
import pandas as pd
```

```python
df = pd.read_csv('breast-cancer.csv')
df.head()
```

```python
df.drop(columns=['id'], inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size = 0.2, random_state = 5)
```

```python
ytrain.shape
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(Xtrain)
X_test = scaler.transform(Xtest)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 10)
```

```python
knn.fit(Xtrain,ytrain)
```

```python
ypred = knn.predict(Xtest)
accuracy_score(ytest,ypred)
```

```python
accuracy = accuracy_score(ytest,ypred)
print(f'ACCURACY = {accuracy:.2%}')
```

```python
# Testing accuracy with multiple KNeighbor value using loop
```

```python
scores = []
for i in range(1,16):
    knn.fit(X_train,ytrain)
```

```python
y_pred = knn.predict(X_test)
scores.append(accuracy_score(ytest,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,16),scores)
plt.show()
```
